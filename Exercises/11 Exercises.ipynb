## Exercise 11.2

For the recursive factorial algorithm in Activity 04, determine the algorithmic complexity by inspecting your implementation of the algorithm. Test this against numerical experiments.

### Solution

Recall the factorial algorithm from Activity 04.4:

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return factorial(n - 1)*n
```

The function calls itself (recursively) $n$ times, hence it has complexity $O(n)$. We test this below and plot
the times.

```python
import time
# Measure time for various n
n_range = list(range(1, 2000, 100))

def complexity_check(func, values = n_range):
    times = []
    for n in values:
        start = time.perf_counter()
        func(n)
        end = time.perf_counter()
        times.append(end - start)
    
    return times

results = complexity_check(factorial)

# Plot results
plt.plot(n_range, results, marker='o')
plt.xlabel('n')
plt.ylabel('Execution time (s)')
plt.title('Runtime of recursive function')
plt.show()
```
